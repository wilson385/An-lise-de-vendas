```python
venda_total = { 'produto': ["Coca-Cola 2,5L", "Bananada", "Pé de moça", "Fofura", "Cerveja Brahma", "Império",
                             "Paçoca", "Guaravita", "Pirulito/G", "Pirulito/P", "Mentos",
                             "Miojo", "Babalu", "Jujuba", "Fini", "Ciclete/ Spich", "Ciclete/blong"],
              'Quantidade': [1, 3, 8, 7, 11, 5, 6, 5, 2, 2, 3, 1, 8, 2, 1, 5, 5],
              'Preço/unitario': [10.00, 0.50, 1.20, 3.50, 6.00, 4.00, 0.70, 2.00, 1.00, 0.50, 1.50, 4.50, 0.50, 0.70, 1.00, 0.25,
                                 0.50]}


vendas_df = pd.DataFrame(venda_total)
display(vendas_df)
```

```python
# Cálculo do total por produto
vendas_df["Total"] = vendas_df["Quantidade"] * vendas_df["Preço/unitario"]

# Cálculo do lucro total
lucro_total = vendas_df["Total"].sum()

# Exibição
display(vendas_df)
display("\nLucro total: R$ {lucro_total:.2f}")
```

```python
print("\nTop 5 produtos com maior faturamento:")
print(vendas_df.sort_values(by="Total", ascending=False).head(5))
```
